```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
fig = plt.figure()

ax = fig.add_axes([0, 0, 1, 1])

x = np.arange(1, 100)
y = 2*x

ax.plot(x, y, 'b')
ax.set_xlim(0, 100)
ax.set_ylim(0, 200)

ax.set_xlabel("X-Axis")
ax.set_ylabel("Y-Axis")
ax.set_title("Title")
```

```python
x = np.arange(1, 100)
y = 2*x

fig = plt.figure()

ax1 = fig.add_axes([0, 0, 1, 1])
ax2 = fig.add_axes([0.2, 0.5, .2, .2])

ax1.plot(x, y, 'r')
ax2.plot(x, y, 'b')

ax1.set_xlim(0, 100)
ax1.set_ylim(0, 200)

ax2.set_xlim(0, 100)
ax2.set_ylim(0, 200)
```

```python
x = np.arange(1, 100)
y = 2*x
z = x*y

fig = plt.figure()

ax1 = fig.add_axes([0, 0, 1, 1])
ax2 = fig.add_axes([0.1, 0.5, .3, .2])

ax1.plot(x, z, 'r')
ax2.plot(x, y, 'b')

ax1.set_xlim(0, 100)
ax1.set_ylim(0, 10000)
ax1.locator_params(axis='both', nbins=6) 

ax2.set_xlim(20, 22)
ax2.set_ylim(30, 50)
ax2.locator_params(axis='both', nbins=5) 

ax1.set_title('Full')
ax1.set_xlabel('X')
ax1.set_ylabel('Z')

ax2.set_xlabel('X')
ax2.set_ylabel('Y')
ax2.set_title('Zoom')
```
